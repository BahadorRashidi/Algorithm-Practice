Write an algorithm that receives two peopole calanders as a list including other lists that shows thestart and end time of each meeting in a day. Also yu will receive another list indicating the available timing of a day for that person.

Try to find all the vailable slots for those two people for 30 mins meeting inthe given day.

```
A = [['9:00','9:50'], ['10:10','10:40'],['12:30','13:45'],['15:00','15:50'],['18:00','18:30']]
A_bound = ['8:00','20:00']
B = [['11:00','11:30'], ['13:00','14:00'],['14:00','14:30'],['17:00','17:30']]
B_bound = ['13:00','18:00']

'''
write a function that convert the strings to the numbers
'''
def convertor(s):
  return int(s[0:s.index(':')])+int(s[s.index(':')+1:])/60

## convert the calender to numebrs
A_c = []
for a in A:
  A_c.append([convertor(a[0]), convertor(a[1])])

B_c = []
for b in B:
  B_c.append([convertor(b[0]), convertor(b[1])])


print('A_c',A_c)
print('B_c',B_c)

## Find the free slots for each calender



A_bound_c = [convertor(A_bound[0]), convertor(A_bound[1])] 
B_bound_c = [convertor(B_bound[0]), convertor(B_bound[1])] 



F_A_c = []
if A_bound_c[0]<A_c[0][0]:
  F_A_c.append([A_bound_c[0],A_c[0][0]])

for i in range(len(A_c)-1):
  if A_c[i][1]>A_bound_c[0] and (A_c[i+1][0] - A_c[i][1])>0:
    F_A_c.append([A_c[i][1], A_c[i+1][0]])

if A_bound_c[1]> A_c[-1][1]:
  F_A_c.append([A_c[-1][1],A_bound_c[1]])
print('F_A_c',F_A_c)


F_B_c = []
if B_bound_c[0]<B_c[0][0]:
  F_B_c.append([B_bound_c[0],B_c[0][0]])

for i in range(len(B_c)-1):
  if B_c[i][1]>B_bound_c[0] and (B_c[i+1][0] - B_c[i][1])>0:
    F_B_c.append([B_c[i][1], B_c[i+1][0]]) 

if B_bound_c[1]> B_c[-1][1]:
  F_B_c.append([B_c[-1][1],B_bound_c[1]])

print('F_B_c', F_B_c)



L=0.5
slot = []
for x in F_A_c:
  for y in F_B_c:
    if (min(x[1],y[1]) - max(x[0], y[0]) >=L):
      slot.append([max(x[0], y[0]),min(x[1],y[1])])

print('slot = ', slot)
```
